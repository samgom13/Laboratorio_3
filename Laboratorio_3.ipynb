## Ejercicio 1
Cree un procedimiento en Python que dado un número entero en decimal despliegue su representación en binario.

```python

def decimal_a_binario(numero: int) -> str:
    """
    Convierte un entero decimal a su representación en binariosin usar bin().
    """
    if numero == 0:
        return "0"

    es_negativo = numero < 0
    numero = abs(numero)
    bits = []

    while numero > 0:
        bits.append(str(numero % 2))
        numero //= 2

    binario = "".join(reversed(bits))
    return "-" + binario if es_negativo else binario


# Pruebas
print(decimal_a_binario(10))   # 1010
print(decimal_a_binario(25))   # 11001
print(decimal_a_binario(0))    # 0
print(decimal_a_binario(-13))  # -1101
```

## Ejercicio 2
Desarrolle una función que devuelva `True` si un número entero es primo y `False` en caso contrario.

```python

def es_primo(numero: int) -> bool:
    """
    Determina si un número entero es primo.
    """
    if numero <= 1:
        return False
    if numero == 2:
        return True
    if numero % 2 == 0:
        return False

    divisor = 3
    while divisor * divisor <= numero:
        if numero % divisor == 0:
            return False
        divisor += 2

    return True


# Pruebas
for n in [2, 3, 4, 5, 9, 11, 25, 29]:
    print(f"{n}: {es_primo(n)}")
```

## Ejercicio 3
Haga un programa que calcule la desviación estándar de una lista de números sin usar funciones que resuelvan directamente el problema.

```python

def desviacion_estandar(lista):
    """
    Calcula la desviación estándar poblacional de una lista de números
    sin usar statistics.stdev() ni numpy.
    """
    n = len(lista)
    if n == 0:
        raise ValueError("La lista no puede estar vacía.")

    suma = 0
    for valor in lista:
        suma += valor
    media = suma / n

    suma_cuadrados = 0
    for valor in lista:
        suma_cuadrados += (valor - media) ** 2

    varianza = suma_cuadrados / n
    return varianza ** 0.5


# Pruebas
datos = [10, 12, 23, 23, 16, 23, 21, 16]
print("Datos:", datos)
print("Desviación estándar:", desviacion_estandar(datos))
```

## Ejercicio 4
Desarrolle una función que ordene un arreglo con el algoritmo de burbuja y devuelva un nuevo arreglo ordenado. Luego cree una función `main` para probarla.

```python

def ordenamiento_burbuja(arreglo):
    """
    Ordena una lista con el algoritmo de burbuja y devuelve
    una nueva lista ordenada.
    """
    resultado = arreglo[:]  # copia
    n = len(resultado)

    for i in range(n):
        for j in range(0, n - i - 1):
            if resultado[j] > resultado[j + 1]:
                resultado[j], resultado[j + 1] = resultado[j + 1], resultado[j]

    return resultado


def main():
    numeros = [64, 34, 25, 12, 22, 11, 90]
    print("Lista original:", numeros)
    print("Lista ordenada:", ordenamiento_burbuja(numeros))


main()
```

## Ejercicio 5
Escriba una función que dada una tupla de elementos (números o letras) la convierta a un `string`.

```python

def tupla_a_cadena(tupla):
    """
    Convierte una tupla de elementos en un string concatenado.
    """
    resultado = ""
    for elemento in tupla:
        resultado += str(elemento)
    return resultado


# Pruebas
entrada_1 = (10, 20, 40, 5, 70)
entrada_2 = ("H", "o", "l", "a")
print(tupla_a_cadena(entrada_1))  # 102040570
print(tupla_a_cadena(entrada_2))  # Hola
```

## Ejercicio 6
Cree un programa que dada una lista de tuplas elimine las tuplas vacías.

```python

def eliminar_tuplas_vacias(lista_de_tuplas):
    """
    Devuelve una nueva lista sin las tuplas vacías.
    """
    resultado = []
    for tupla in lista_de_tuplas:
        if tupla != ():
            resultado.append(tupla)
    return resultado


# Pruebas
lista = [(), (), ('X',), ('a', 'b'), ('a', 'b', 'c'), ('d',)]
print("Original:", lista)
print("Sin vacías:", eliminar_tuplas_vacias(lista))
```

## Ejercicio 7
Cree una función que dada una tupla de tuplas con números produzca una tupla con el promedio de cada tupla individual.

```python

def promedio_por_tupla(tupla_de_tuplas):
    """
    Calcula el promedio de cada tupla interna y devuelve una nueva tupla.
    """
    promedios = []

    for subtupla in tupla_de_tuplas:
        if len(subtupla) == 0:
            raise ValueError("No se puede calcular el promedio de una tupla vacía.")

        suma = 0
        for numero in subtupla:
            suma += numero

        promedios.append(suma / len(subtupla))

    return tuple(promedios)


# Pruebas
datos = ((10, 10, 10, 12), (30, 45, 56, 45), (81, 80, 39, 32))
print(promedio_por_tupla(datos))  # (10.5, 44.0, 58.0)
```

## Ejercicio 8
Crear una clase `Libro` con atributos para título, autor, número de páginas y si ha sido leído o no.

```python

class Libro:
    def __init__(self, titulo, autor, numero_paginas, leido=False):
        self.titulo = titulo
        self.autor = autor
        self.numero_paginas = numero_paginas
        self.leido = leido

    def marcar_como_leido(self):
        self.leido = True

    def descripcion(self):
        estado = "Sí" if self.leido else "No"
        return (
            f"Título: {self.titulo}\n"
            f"Autor: {self.autor}\n"
            f"Número de páginas: {self.numero_paginas}\n"
            f"Leído: {estado}"
        )


# Pruebas
libro = Libro("Cien años de soledad", "Gabriel García Márquez", 417)
print(libro.descripcion())
print("\nMarcando como leído...\n")
libro.marcar_como_leido()
print(libro.descripcion())
```

## Ejercicio 9
Crear una clase `Circulo` con métodos para calcular área y circunferencia.

```python

class Circulo:
    def __init__(self, radio):
        if radio < 0:
            raise ValueError("El radio no puede ser negativo.")
        self.radio = radio

    def area(self):
        return 3.141592653589793 * (self.radio ** 2)

    def circunferencia(self):
        return 2 * 3.141592653589793 * self.radio


# Pruebas
circulo = Circulo(5)
print("Radio:", circulo.radio)
print("Área:", circulo.area())
print("Circunferencia:", circulo.circunferencia())
```
